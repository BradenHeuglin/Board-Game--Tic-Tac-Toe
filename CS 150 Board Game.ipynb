```python
import pygame
import sys
import time
pygame.init()

WIDTH = 400
HEIGHT=400
surface = pygame.display.set_mode((WIDTH, HEIGHT+ 100), 0, 32)
pygame.display.set_caption("TicTacToe")

WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
GRAY = (200, 200, 200)
RED = (255, 0, 0)
BLUE = (0, 0, 255)
GREEN=(0,255,0)

welcome = pygame.transform.scale(pygame.image.load("welcome.jfif"), (WIDTH, HEIGHT+100))
x_img = pygame.transform.scale(pygame.image.load("cross.png"), (100, 100))
o_img = pygame.transform.scale(pygame.image.load("circle.png"), (100,100))

FONT =pygame.font.Font('freesansbold.ttf', 30)

winner = None
 
draw = None

XO = 'x'

fps = 30

CLOCK = pygame.time.Clock()

board = [[None]*3, [None]*3, [None]*3]

def game_initiating_window():

    surface.blit(welcome, (0, 0))

    pygame.display.update()
    time.sleep(3)                   
    surface.fill(WHITE)
  
    pygame.draw.line(surface, GRAY, 
                     (WIDTH / 3, 0), 
                     (WIDTH / 3, HEIGHT), 7)
    pygame.draw.line(surface, GRAY, 
                     (WIDTH / 3 * 2, 0), 
                     (WIDTH / 3 * 2, HEIGHT), 7)
  
    pygame.draw.line(surface, GRAY, 
                     (0, HEIGHT / 3), 
                     (WIDTH, HEIGHT / 3), 7)
    pygame.draw.line(surface, GRAY, 
                     (0, HEIGHT / 3 * 2), 
                     (WIDTH, HEIGHT / 3 * 2), 7)
    draw_status()
  
def draw_status():
     
    global draw, FONT
     
    if winner is None:
        message = XO.upper() + "'s Turn"
    if winner:
        message = winner.upper() + " won !"
        
    if draw:
        message = "Game Draw !"

    text = FONT.render(message, 1, WHITE)
  
    surface.fill (BLACK, (0, 400, 500, 100))
    text_rect = text.get_rect(center =(WIDTH / 2, 500-50))
    surface.blit(text, text_rect)
    pygame.display.update()
     
def check_win():
    global board, winner, draw
  
    for row in range(0, 3):
        if((board[row][0] == board[row][1] == board[row][2]) \
           and (board [row][0] is not None)):
            winner = board[row][0]
            pygame.draw.line(surface, GREEN,
                         (0, (row + 1)*HEIGHT / 3 -HEIGHT / 6),\
                         (WIDTH, (row + 1)*HEIGHT / 3 - HEIGHT / 6 ), 4)
            break
  
    for col in range(0, 3):
        if((board[0][col] == board[1][col] == board[2][col]) 
           and (board[0][col] is not None)):
            winner = board[0][col]
            pygame.draw.line (surface, GREEN, 
                              ((col + 1)* WIDTH / 3 - WIDTH / 6, 0), \
                          ((col + 1)* WIDTH / 3 - WIDTH / 6, HEIGHT), 4)
            break
  
    if (board[0][0] == board[1][1] == board[2][2]) \
    and (board[0][0] is not None):
         
        winner = board[0][0]
        pygame.draw.line (surface , GREEN, (10, 10), (390, 390), 4)
         
    if (board[0][2] == board[1][1] == board[2][0]) \
    and (board[0][2] is not None):
         
        winner = board[0][2]
        pygame.draw.line (surface, GREEN, (390, 10), (10, 390), 4)
  
    if(all([all(row) for row in board]) and winner is None ):
        draw = True
    draw_status()
     
def drawXO(row, col):
    global board, XO
      
    if row == 1:
        posx = 15
            
    if row == 2:
 
        posx = WIDTH / 3 + 15
         
    if row == 3:
        posx = WIDTH / 3 * 2 + 15
  
    if col == 1:
        posy = 15
         
    if col == 2:
        posy = HEIGHT / 3 + 15
     
    if col == 3:
        posy = HEIGHT / 3 * 2 + 15
         
    board[row-1][col-1] = XO
     
    if(XO == 'x'):
         
        surface.blit(x_img, (posy, posx))
        XO = 'o'
     
    else:
        surface.blit(o_img, (posy, posx))
        XO = 'x'
    pygame.display.update()
  
def user_click():
    
    x, y = pygame.mouse.get_pos()
  
    if(x<WIDTH / 3):
        col = 1
     
    elif (x<WIDTH / 3 * 2):
        col = 2
     
    elif(x<WIDTH):
        col = 3
     
    else:
        col = None
  

    if(y<HEIGHT / 3):
        row = 1
     
    elif (y<HEIGHT / 3 * 2):
        row = 2
     
    elif(y<HEIGHT):
        row = 3
     
    else:
        row = None
       
    if(row and col and board[row-1][col-1] is None):
        global XO
        drawXO(row, col)
        check_win()
         
def reset_game():
    global board, winner, XO, draw
    time.sleep(3)
    XO = 'x'
    draw = False
    game_initiating_window()
    winner = None

    board = [[None]*3, [None]*3, [None]*3]
   
    message = XO.upper() + "'s Turn"
    
    text = FONT.render(message, 1, WHITE)
    surface.fill (BLACK, (0, 400, 500, 100))
    text_rect = text.get_rect(center =(WIDTH / 2, 500-50))
    surface.blit(text, text_rect)
    pygame.display.update()
  
game_initiating_window()

active=True
while(active):
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            active=False
            pygame.quit()
            sys.exit()
        elif event.type== pygame.MOUSEBUTTONDOWN:
            user_click()
            if(winner or draw):
                reset_game()
    pygame.display.update()
    CLOCK.tick(fps)
```
